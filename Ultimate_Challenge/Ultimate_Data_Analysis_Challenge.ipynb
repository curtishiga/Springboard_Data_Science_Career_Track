# Part 1 - Exploratory Data Analysis

### Problem:
From the *logins.json*:
+ Aggregate login counts based on 15-minute time intervals
+ Visualize and describe the resulting time series of login counts in ways that best characterize the uderlying patterns of the demand
+ Please report/illustrate important features of the demand, such as daily cycles
+ Report data quality issues

After first taking a look at the format of the data, I'll tackle this problem by importing the data and setting the index as the dates. That'll make resampling it into 15 minute intervals and taking the counts easier. It may also be useful to add which day of the week each date is. First, a check needs to be done to see if there are any duplicate values or times where more than one login was registered at the same time.

```python
# Import the json file as a pandas DataFrame
with open('logins.json', 'r') as json_file:
    json_data = json.load(json_file)

# Create a dataframe with login_time as the index
logins = pd.DataFrame(json_data)
```

```python
logins.head()
```

```python
# Get number of total values
len(logins)
```

```python
# Get number of unique values
logins.login_time.nunique()
```

Seeing as there is a difference in the number of total values and the number of unique values, there must be duplicate values that need to be accounted for. One way to get around this is to group by the timestamp and aggregate it by count.

```python
# Create a column to count/sum up
logins['count'] = 1

logins_grouped = logins.groupby('login_time').count()
```

```python
logins_grouped.head()
```

```python
# Convert index into datetime objects
logins_grouped.index = pd.to_datetime(logins_grouped.index)
```

```python
logins_grouped.head()
```

```python
# Resample the dataa by 15 minute intervals and take the count
logins_resampled = logins_grouped.resample('15T').sum()
```

```python
# Add a column for day of week
logins_resampled['day_of_week'] = logins_resampled.index.dayofweek
```

```python
logins_resampled.head()
```

```python
# Plot a plot of the distribution
# Histogram takes too long to render with so many bins
_ = plt.figure(figsize=(15,15))
_ = plt.plot(logins_resampled['count'])
_ = plt.title('Number of Logins over Time')
_ = plt.xlabel('Time')
_ = plt.ylabel('Number of Logins')
```

There seems to be a regularity in the time of logins. They could depend on time of day, day of week, or even day of month. Further plots depending on each of these variables should be made to help visualize the data.

```python
# Group by day, time of day (hour), and day of week
grouped_day = logins_resampled.groupby(logins_resampled.index.day).sum()
grouped_hr = logins_resampled.groupby(logins_resampled.index.hour).sum()
grouped_dow = logins_resampled.groupby(logins_resampled.day_of_week).sum()
```

```python
grouped_day.head()
```

```python
# Plot histogram by day, hour, and day of week
_ = plt.figure(figsize=(10,10))
_ = plt.plot(grouped_day['count'])
_ = plt.title('Number of Logins by Day of Month')
_ = plt.xlabel('Day of Month')
_ = plt.ylabel('Number of Logins')
```

```python
_ = plt.figure(figsize=(10,10))
_ = plt.plot(grouped_hr['count'])
_ = plt.title('Number of Logins by Hour')
_ = plt.xlabel('Hour')
_ = plt.ylabel('Number of Logins')
```

```python
_ = plt.figure(figsize=(10,10))
_ = plt.plot(grouped_dow['count'])
_ = plt.title('Number of Logins by Day of Week')
_ = plt.xlabel('Day of Week')
# _ = plt.xticks(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
_ = plt.ylabel('Number of Logins')
```

There doesn't seem to be any conclusive trends when looking at the time of the month. The first of three plots above shows no significant difference in logins as the month progresses. The following two plots show more distinctive patterns in regards to time of day and day of the week. In the second plot, there are spikes in the number of logins that show midday and midnight activity. In the last plot, activity seems to be minimal on Monday's but as the week goes on, the number of logins daily gradually increases peaking around Saturday.
